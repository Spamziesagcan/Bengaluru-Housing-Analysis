```python
df[[ 'marks_subject1', 'marks_subject2', 'marks_subject3']].mean()
```

8. Add a New Column: Add a new column 'total_marks' which is the sum of marks in all three subjects for each student.

```python
df['total_marks'] = df[[ 'marks_subject1', 'marks_subject2', 'marks_subject3']].apply(lambda x: sum(x),axis=1)
df
```

9. Sort by Total Marks: Sort the DataFrame based on 'total_marks' in descending order.

```python
df.groupby('department')[["marks_subject1","marks_subject2","marks_subject3"]].sum()
```

18. Normalize Marks: Create a new column 'normalized_marks' which is the marks in 'marks_subject1' normalized to a range of 0 to 1.

```python
df['normalized_marks'] = df[['marks_subject1']].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
df
```
